```python
from pymystem3 import Mystem
import string
import pandas as pd
import numpy as np
import re
import os
```

```python
def letters(input):
    return ''.join(filter(lambda x: x.isalpha() or x == '-', input))     
```

```python
with open(r"C:\Users\Admin\Desktop\Практика Влада\years.txt", "r") as f:
    years = f.read().split()
```

```python
path_to_txts = r"C:\Users\Admin\Desktop\Практика Влада\Статьи в txt"
words_dict = {}
clean = []

# for i in len(os.listdir(path_to_txts)):
for i in range(10):
    file = r"C:\Users\Admin\Desktop\Практика Влада\Статьи в txt\{}.txt".format(i)
    with open(file, 'r', encoding='utf-8') as f:
        for line in f.readlines():
            clean += [letters(word) for word in line.split()]
        if years[i] in words_dict.keys():
            words_dict[years[i]] = words_dict[years[i]] + list(filter(lambda x: x != '' and len(x) > 1, clean))
        else:
            words_dict[years[i]] = list(filter(lambda x: x != '' and len(x) > 1, clean))
        clean = []
print(words_dict)
```

```python
m = Mystem()
all_words = ""
eng_letters = [x for x in string.ascii_letters]

for key in words_dict:
    for i, word in enumerate(words_dict[key]):
        if word[-1] == '-':
            words_dict[key][i] = word.replace("-", "") + words_dict[key][i+1]
            del words_dict[key][i+1]
        if "cid" in word:
            words_dict[key][i] = word.replace("cid", "") + words_dict[key][i+1]
            del words_dict[key][i+1]
        if isinstance(word, list):
            words_dict[key][i] = word[0]
        if word[0] in eng_letters:
             del words_dict[key][i]
        else:    
            all_words += words_dict[key][i] + " "
    
    all_words = list(filter(lambda x: x != ' ', m.lemmatize(all_words)))
    words_dict[key] = all_words
    all_words = ""

print(words_dict['2016'])
```

```python
print("Введите слово или словосочетание") # максимум два слова
user_input = list(filter(lambda x: x != ' ', m.lemmatize(input())))
user_input
```

```python
count, total = 0, 0
output = {}
for key in words_dict:
    for i, word in enumerate(words_dict[key]):
        if user_input[0] == word:
            if len(user_input) > 1 and i+1 != len(words_dict[key]) and user_input[1] == words_dict[key][i+1]:
                count += 1
            if len(user_input) == 2:
                count += 1
    output[key] = count
    total += count
    count = 0
```

```python
output_years = list(map(int, output.keys()))
output_years.sort()

for year in output_years:
    print("{}: {}".format(year, output[str(year)]))  
```
