```python
def check_containts(source: str, target: list, return_bool=False):
    if type(source) != str:
        if return_bool:
            return False

        return set()

    source = source.lower().split("; ")
    contains = set()
    for t in target:
        if t in source:
            if return_bool:
                return True

            contains.add(t)

    if return_bool:
        return False
    return contains


def keywords_by_year(source_df, source_col="Ключевые слова указателя"):
    year_to_words = defaultdict(list)
    #     stopwords = get_stop_words(1)
    for words_text, year in tqdm(source_df[[source_col, "Год"]].values):
        intersections = check_containts(words_text, keywords["keywords"])
        if not intersections:
            continue

        for word in intersections:
            year_to_words[year].append(word)

    return year_to_words


def get_stop_words(treshold):
    c = Counter(
        "; ".join(data_health["Ключевые слова указателя"].values).lower().split("; ")
    )
    stopwords = [
        i[0] for i in c.most_common() if i[1] / data_health.shape[0] < treshold
    ]
    return stopwords


def year_top_words(year, n):
    words = year_to_words[year]
    c = Counter(words)
    top_words = pd.DataFrame(c.most_common(n), columns=("word", "count"))
    sns.barplot(data=top_words, x="count", y="word")
    plt.title(year)


def filter_by_word(source_df, col_name, word):
    res = source_df[source_df[col_name].astype(str).apply(lambda x: word in x.lower())]
    return res
```

# Работа с кратким содержанием

```python
import string
import spacy
import re
```

```python
nlp = spacy.load("en_core_web_lg")
```

```python
STOPWORDS = spacy.lang.en.STOP_WORDS
```

```python
def preprocess(text):
    if type(text) != str:
        return []

    regx = re.compile("\W")
    text = regx.sub(" ", text)
    doc = nlp(text)
    tokens = []
    noun_chunks = []
    for token in doc:
        lemma = token.lemma_
        if lemma.strip() == "" or len(lemma) == 1:
            continue

        if lemma not in string.punctuation and lemma not in STOPWORDS:
            tokens.append(lemma)

    for nc in doc.noun_chunks:
        nc = nc.text.lower()
        if nc not in string.punctuation and nc not in STOPWORDS:
            noun_chunks.append(nc.strip())

    return tokens, noun_chunks
```
